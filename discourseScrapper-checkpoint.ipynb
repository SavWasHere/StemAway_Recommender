```python
import requests
from bs4 import BeautifulSoup
import csv
import re
import random

from selenium import webdriver
from selenium.webdriver.chrome.service import Service



class formNavigator():
    
    def __init__ (self, basicLink, websitePage="", specificClasses = [], specificTags = [], specificIds = [], blackList = {}):
        
    
        self.basicLink = basicLink
        self.websitePage = websitePage
        
        self.linksStored = {set} 
        self.blackList = {set}
        
        self.bSoup = None
        self.driver = None

        self.pageType = None
    
    def getPageHtml(self):
        
        
        selService = Service('C:/Users/dilsh/Downloads/chromedriver_win32/chromedriver.exe')
        
        option = webdriver.ChromeOptions()
        option.add_argument('headless')
        
        self.driver = webdriver.Chrome(service=selService, options=option)
        self.driver.get(f'{self.basicLink}')
        self.driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        
        with open("htmlStored.txt", "r+", encoding='utf-8') as htmlPage:
            htmlPage.write(self.driver.page_source) 
            
        self.bSoup = BeautifulSoup(self.driver.page_source, 'html.parser')
        return self.bSoup
            
    
    def getLink(self):
        
        for aTag in self.bSoup.find_all("a"):
            
            try:
                aTag = (aTag.get("href"))


                if aTag not in self.blackList and aTag.startswith("/") and len(aTag) > 3 :
                    self.linksStored.add(aTag)
            
            except AttributeError as a:
                continue
            
                
    
    def setLink(self):
    
        nextPage = random.choice(tuple(self.linksStored))
        
        if "/c/" in nextPage:
            driver.get(self.basicLink+nextPage)
            self.pageType = "category"
            
            
        elif "/u/" in nextPage:
            driver.get(self.basicLink+nextPage)
            self.pageType = "userPage"
        
        
        elif "/t/" in nextPage:
            driver.get(self.basicLink+nextPage)
            self.pageType = "formPost"
        
        else:
            driver.get(self.basicLink+nextPage)
            self.pageType = "miscellaneous"
        
        
        
        return self.pageType
        
```
